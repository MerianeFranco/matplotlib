1 - There is a positive and large association between 
 the tumor volume and the weight of mice, indicated by the Pearson correlation coefficient equal to 0.951.
 
2 - The drugs Capomulin and Remicare presented less tumor volume and were the drugs that has highes number of measurements.

3 - The gender distribution on the study was balanced, 49.6% Female and 50.4% Male mice.

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
from scipy.stats import linregress

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
mouse_df = pd.merge(mouse_metadata,study_results)

# Display the data table for preview
mouse_df.head()
```

```python
# Checking the number of mice.
len(mouse_df['Mouse ID'].unique())
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint.
# Part 1 - Concatenate values "Mouse ID" and "Timepoint" in a new column
mouse_df["MouseTime"]=mouse_df["Mouse ID"].map(str) +'-'+mouse_df["Timepoint"].map(str)
# Part 2 - Create a new column to check if "Mouse ID" and "Timepoint" are duplicated
mouse_df['Repeated']=mouse_df.duplicated(subset=['MouseTime'])
mouse_df.head()
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint.
# Part 3 - looking for the duplicate mouse in all rows
i =[]
for row in range(len(mouse_df)):
    if mouse_df.iloc[row,9] == True:
        i.append(mouse_df.iloc[row,0])
pd.Series(i).unique()

```

```python
# Optional: Get all the data for the duplicate mouse ID.
mouseID_df = mouse_df.set_index("Mouse ID")
mouseID_df.loc['g989',:]
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
Cleaning_df = mouseID_df.drop(index=['g989'])
Cleaning_df = Cleaning_df.drop(columns=['MouseTime','Repeated'])

Clean_df = Cleaning_df.reset_index()
Clean_df.head()
```

```python
# Checking the number of mice in the clean DataFrame.
len(Clean_df['Mouse ID'].unique())
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 

#Grouping
grouped = Clean_df.groupby("Drug Regimen")

#Calculate Values
#Mean
mean = grouped["Tumor Volume (mm3)"].mean()
#Median
median = grouped["Tumor Volume (mm3)"].median()
#Variance
variance = grouped["Tumor Volume (mm3)"].var()
#Standard Deviation
std = grouped["Tumor Volume (mm3)"].std()
#Standard Error of the Mean (SEM)
sem = grouped["Tumor Volume (mm3)"].sem()

# Assemble the resulting series into a single summary dataframe.
summary_df = pd.DataFrame({"Mean Value":mean,
                          "Median Value":median,
                          "Variance":variance,
                          "Standard Deviation":std,
                          "SEM":sem})
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
summary_df
```

```python
# Using the aggregation method, produce the same summary statistics in a single line
summary2_df = grouped.agg({"Tumor Volume (mm3)":["mean","median","var","std","sem"]})
summary2_df
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of 
#the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse

NewClean_df = Clean_df.groupby("Mouse ID").max()["Timepoint"]
NewClean_df = NewClean_df.reset_index()
NewClean_df

# Merge this group df with the original dataframe to get the
#tumor volume at the last timepoint
Merge_df = pd.merge(NewClean_df,Clean_df, on = ["Mouse ID","Timepoint"])
Merge_df.head()
```

```python
# Put treatments into a list for for loop (and later for plot labels)
treatment = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]

# Create empty list to fill with tumor vol data (for plotting)
tumorvolume = []

# Locate the rows which contain mice on each drug and get the tumor volumes
for treat in treatment:
    
    Tumordata = Merge_df.loc[Merge_df["Drug Regimen"]==treat]["Tumor Volume (mm3)"]   
    tumorvolume.append(Tumordata)

# Calculate the IQR and quantitatively determine if there are any potential outliers. 
# add subset 
    quartiles = Tumordata.quantile([.25,.5,.75])
    lowerq = quartiles[0.25]
    upperq = quartiles[0.75]
    iqr = upperq-lowerq

# Determine outliers using upper and lower bounds
    lower_bound = lowerq - (1.5*iqr)
    upper_bound = upperq + (1.5*iqr)

    outlier = Tumordata.loc[(Tumordata < lower_bound)|(Tumordata > upper_bound)]

    print(treat, outlier)
```

```python
# Generate a box plot of the final tumor volume of each mouse across four regimens of interest
red_diamond = dict(markerfacecolor='r', marker='D')
plt.title('Tumor Volume (mm3) per treatment')
plt.ylabel('Tumor Volume (mm3)')
plt.boxplot(tumorvolume, labels = treatment, flierprops = red_diamond)
plt.show()
```

## Line and Scatter Plots

```python
# Generate a scatter plot of average tumor volume vs. mouse weight for the Capomulin regimen

#Getting the DataFrame for Capomulin Regimen, grouped by weight
Capomulin_df = Clean_df.loc[Clean_df["Drug Regimen"]=='Capomulin']
grouped_weight = Capomulin_df.groupby("Weight (g)")

#Calculate the average for tumor volume
mean = grouped_weight["Tumor Volume (mm3)"].mean()
mean_df = pd.DataFrame(grouped_weight["Tumor Volume (mm3)"].mean())
mean_df

# generating the scatterplot
smean_df = Capomulin_df.groupby("Mouse ID").mean()

x_axis=smean_df["Weight (g)"]
y_axis=smean_df["Tumor Volume (mm3)"]

plt.title("Average tumor volume vs. mouse weight for Capomulin Drug Regimen")
plt.xlabel("Mouse weight")
plt.ylabel("Average tumor volume")
plt.scatter(x_axis, y_axis, marker="o", facecolors="red", edgecolors="black",
             alpha=0.75)
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen

weight = mean_df.index
average_tumor = mean_df['Tumor Volume (mm3)'].tolist()
correlation = st.pearsonr(weight,average_tumor)
print(f"The correlation between both factors is {round(correlation[0],3)}")
```

```python
#linear regression model 
x_axis=smean_df["Weight (g)"]
y_axis=smean_df["Tumor Volume (mm3)"]
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_axis, y_axis)
regress_values = x_axis * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))

plt.annotate(line_eq,(22,38),fontsize=10,color="red")

plt.title("Average tumor volume vs. mouse weight for Capomulin Drug Regimen")
plt.xlabel("Mouse weight")
plt.ylabel("Average tumor volume")
plt.scatter(x_axis, y_axis, marker="o", facecolors="blue", edgecolors="black",
             alpha=0.90)
plt.plot(x_axis,regress_values,"r-")
print(f"The r-squared is: {round(rvalue**2,5)}")
plt.show()
```
